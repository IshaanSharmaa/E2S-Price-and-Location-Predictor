```python
dff = df1.drop(['Seller ProfileID ', 'Seller Company Name', "Buyer Company Name", "Buyer Profile Id"], axis = 1 )
```

```python
dff.columns
```

```python
x = dff.drop(['Buyer State', 'Buyer Location', 'Rate'], axis = 1)
x.head()
```

```python
y = dff[['Buyer State', 'Buyer Location', 'Rate']]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder


column_to_be_encoded = [
      'Seller State', 'Seller Location', 'Product Group', 'Product Category',
      'Product Subcategory', 'Product Offered'
      ]

encoder = OrdinalEncoder()
encoding_transformer = ColumnTransformer(
      [
              ("encod", encoder, column_to_be_encoded)
      ],
      remainder= "passthrough"       
)

encoding_transformer.fit(x)
transformed_x_np = encoding_transformer.transform(x)
transformed_x = pd.DataFrame(transformed_x_np, columns = ['Seller State', 'Seller Location', 'Product Group', 'Product Category',
'Product Subcategory', 'Product Offered', 'Qty', 'Actual Price',
'Seller Price'])
# return transformed_x

encoded_x = transformed_x
```

```python
encoded_x
```

FOR PRICE

```python
x_price = encoded_x.drop(['Seller State', 'Seller Location'], axis = 1)
y_price = dff[['Rate']]
len(x), len(y_price), len(y_price)
```

```python
x_price.columns
```

```python
from sklearn.model_selection import train_test_split

xp_train, xp_test, yp_train, yp_test = train_test_split(x_price, y_price, test_size= 0.2)

len(xp_train), len(yp_train), len(xp_test), len(yp_test)
```

```python
'''
Random Forest Regressor
Ridge Regerssion
SVR kernel = linear
'''

np.random.seed(100)

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge

model_dict = {
    "RandomForest" : RandomForestRegressor(),
    "Ridge" : Ridge()
}

score = {}

for model_name, model in model_dict.items():

    print("Fitting: ", model_name)
    model.fit(xp_train, np.ravel(yp_train, order = "C"))           

    score[model_name] = model.score(xp_test, yp_test)
```

```python
score
```

```python
np.random.seed(100)

model_p = Ridge()
model_p.fit(xp_train, np.ravel(yp_train))
model_p.score(xp_test, yp_test)
```
